## Research Question

Among higher-education students, does a country’s economic development level relate to (a) how frequently students use ChatGPT for schoolwork and (b) the main way they use it—learning support (tutoring/explaining/summarizing) versus assignment production (drafting/writing/coding/generating deliverables)?


## Background and Prior Work


AI has the potential to improve a wide range of fields and these industries may vary depending on the country. For example, in Kenya, AI is utilized to analyze the soil and crop conditions in agriculture.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) On the other hand, the healthcare industry in South Africa utilizes AI to better understand and meet patient needs.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) AI powered tools have substantially reduced human errors in such fields which require precise calculations, leading to a rapid development of new technology, thus enhancing a country’s economic growth. Vice versa, the economic state of a country has the potential to shape how AI is used.

While AI has been a tool to improve the economic state of a country, it also has improved our daily lives by conveniently summarizing information at the tips of our fingers. With an increase in AI investment contributing significantly to the GDP growth since 2014<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2), a majority of students from various fields, surveyed between late 2023 and early 2024, believe that using AI improves learning and assignment execution in general<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). However, a portion of the surveyed students expressed concerns over AI’s reliability and ethicalness within the academic setting.

Despite its benefit for the overall global development, AI has the potential to backfire. With an increase in AI usage comes an increase in concerns over how AI is utilized as a tool. In prior research and analysis, results reflected the trade off of using generative AI, a declination in critical thinking. In the study conducted by Kosmyna et al<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4), essays produced by participants in the LLM group reflected weak variability compared to the participants in the Brain-only group and the Search Engine group.

With the variables of economic utilizations of AI and AI’s effects on human cognition, does a country’s development determine AI usage, especially in students? Specifically, does a country’s development relate to student usage of AI, the ways in which it is used as well as how much it is used?

## Hypothesis


Hypothesis: Country economic development will predict a dual pattern in ChatGPT usage: higher-income countries will show both greater usage frequency and higher allocation to learning support (tutoring/explanations/summaries), while lower-income countries will show lower frequency but proportionally more assignment production (drafting/writing/coding/generating deliverables).



Explanation:We predict that higher-education students in more economically developed countries will report using ChatGPT more often for schoolwork. We also predict that, among users, students in less developed countries will use ChatGPT relatively more for assignment production (drafting/writing/coding), while students in more developed countries will use it relatively more for learning support (tutoring/explanations/summaries). We expect this pattern because access and adoption tend to be higher in wealthier countries, while higher academic pressure and fewer support resources may push students in lower-income settings toward using ChatGPT to generate deliverables, and stricter academic integrity norms in some higher-income settings may discourage production-style use.


## Data

The ideal dataset we would expect to have would combine student level survey data with country level economic indicators.

At the country level, the dataset would include measures of economic development such as GDP per capita, World Bank income classification, and internet access rates. These indicators would be used to represent a country’s level of economic development and would be merged with student responses using standardized country codes. 

At the student level, the dataset would collect basic demographic and academic information, including age, level of study, field of study, primary language, and self-reported English proficiency, which are important for controlling differences in access to and reliance on ChatGPT. The dataset would also include detailed measures of ChatGPT usage for schoolwork. This would include how frequently students use ChatGPT, measured on a scale ranging from never to daily use, as well as questions about how they primarily use it. Students would report whether they mainly use ChatGPT for learning support (such as explaining concepts, summarizing readings, or studying) or for assignment production (such as drafting essays, writing code, or generating deliverables). Ideally, data would be collected from several thousand higher-education students across multiple countries through an anonymous online survey distributed via universities and student organizations. 

The final dataset would be organized in a tidy format, with one row per student and country-level variables merged in, so it would be easier for clear cross-country comparisons and statistical analysis.

When analyzing the data, it’s important to consider both direct and indirect factors that could affect how students use ChatGPT. While a country’s level of economic development is the main focus, other variables like field of study, language background, and access to technology could influence ChatGPT usage at the same time. If these factors are ignored, the results could be misleading. Including them helps reduce confounding and makes the findings more applicable across different student groups. 

The dataset is also checked for potential bias. Since the data are self-reported, the same students may not report their ChatGPT usage accurately, either because they are unsure how to categorize their use or because they are worried about academic consequences. Some countries may also have fewer responses than others, which could skew the comparisons. These issues are addressed by checking for imbalance groups, avoiding strong conclusions from small samples, and being transparent about uncertainty in the results. Visualizations and summary statistics are kept simple and clearly labeled so the data are not exaggerated or misrepresented. 

Student privacy is protected throughout the analysis by not collecting or displaying any personally identifiable information. All results are reported at an aggregate level, and no individual student responses are singled out. Finally, the analysis process is documented step by step, including how variables are created and how decisions are made, so the work can be reviewed or repeated later if problems are found. This helps keep the analysis transparent and accountable.
